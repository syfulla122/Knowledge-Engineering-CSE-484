```python
#import libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
```

```python
#read dataset
df = pd.read_csv('/content/drive/MyDrive/new drive/Knowledge engineering/titanic.csv')
df.head()
```

```python
#check for missing values
print ("missing values before handling")
df.isnull().sum()
```

```python
#handling missing values
#1.Fill missing age with median age
df['Age'].fillna(df['Age'].median(), inplace=True)
#2.Fill missing embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
#3.drop the cabin column
df.drop('Cabin', axis=1, inplace=True)
#4.check missing values after handling
print ("missing values after handling")
df.isnull().sum()
```

```python
#bar chart-survival distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.xticks([0,1], ['Not Survived', 'Survived'])
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distribution on Titanic')
plt.show()
```

```python
#bar chart - survival by gender
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set2')
plt.title('Survival by Gender')
plt.legend(['Not Survived', 'Survived'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#histogram - age distribution
plt.figure(figsize=(6,4))
sns.distplot(df['Age'].dropna(), bins=30,kde=True,color ="blue")
plt.title('Age Distribution of passanger')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#box plot - fare vs survival
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set2')

plt.title('Fare vs Survival')
plt.xlabel('Survival status')
plt.ylabel('Fare')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Prepare the data
# Use the original df DataFrame and handle 'Embarked'
le = LabelEncoder()
df['Embarked_encoded'] = le.fit_transform(df['Embarked'])

# Select features (excluding 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'PassengerId') and target
features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
target = 'Embarked_encoded'

X = df[features]
y = df[target]

# Handle potential missing values in features (if any, although we handled them before)
X = X.fillna(X.median())

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features) # Convert back to DataFrame with column names

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model Selection (Logistic Regression)
model = LogisticRegression(max_iter=1000) # Increased max_iter for convergence

# Model Training
model.fit(X_train, y_train)

# Model Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the Logistic Regression model: {accuracy:.4f}")

# Display the first few predictions and actual values
print("\nFirst 10 predictions vs actual:")
for i in range(10):
    print(f"Predicted: {le.inverse_transform([y_pred[i]])[0]}, Actual: {le.inverse_transform([y_test.iloc[i]])[0]}")

# Print classification report for precision, recall, and F1-score
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0))
```
