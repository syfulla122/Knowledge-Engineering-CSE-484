```python
#import libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#directed graph
G = nx.DiGraph()

#add nodes
G.add_nodes_from(["A","B","C","D"])

#add edges
G.add_edges_from([("A","B"),("B","C"),("C","A"),("A","D")])

#draw graph
nx.draw(G,with_labels=True,node_color='lightblue')
```

```python
#in degree
print("In degree of each node:")
for node,deg in G.in_degree():
  print(f"{node}:{deg}")

#out degree
print("Out degree of each node:")
for node,deg in G.out_degree():
  print(f"{node}:{deg}")

#total degree
print("Total degree:")
for node,deg in G.degree():
  print(f"{node}:{deg}")
```

```python
node = "A"
print(f"successor of  {node}:",list(G.successors(node)))
print(f"predecessor of  {node}:",list(G.predecessors(node)))
```

```python
node = "D"
print(f"successor of  {node}:",list(G.successors(node)))
print(f"predecessor of  {node}:",list(G.predecessors(node)))
```

```python
#successor of nodes
print("Successor of nodes:")
for node in G.nodes():
  print(f"{node}:{list(G.successors(node))}")
#predessor of node
print("Predessor of nodes:")
for node in G.nodes():
  print(f"{node}:{list(G.predecessors(node))}")
```

```python
#adjacency matrix of my graph
adj_matrix = nx.adjacency_matrix(G)
print("Adjacency matrix:")
print(adj_matrix.todense())
#draw adjacency matrix
plt.matshow(adj_matrix.todense())
plt.show()
```

```python
import pandas as pd
adj_df = pd.DataFrame()
adj_df = pd.DataFrame(nx.adjacency_matrix(G),index=sorted(G.nodes()),columns=sorted(G.nodes()))
print(adj_df)
```

```python
#sortest path finding
print(nx.shortest_path(G,"A","C"))
print(nx.shortest_path_length(G,"A","C"))
```

```python
#find cycle
print(nx.find_cycle(G))
```

```python
#strongly connected component
print(nx.is_strongly_connected(G))
print(list(nx.strongly_connected_components(G)))
```
